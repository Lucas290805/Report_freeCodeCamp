```python
my_graph = {
    'A': [('B', 5), ('C', 3), ('E', 11)],
    'B': [('A', 5), ('C', 1), ('F', 2)],
    'C': [('A', 3), ('B', 1), ('D', 1), ('E', 5)],
    'D': [('C',1 ), ('E', 9), ('F', 3)],
    'E': [('A', 11), ('C', 5), ('D', 9)],
    'F': [('B', 2), ('D', 3)]
}
```

```python
def shortest_path(graph, start, target = ''):

# Tao 2 list unvisited: luu cac va quan li cac diem da di qua
    unvisited = list(graph)
#vd: khoang cach tai A la 0, cac diem chua xac dinh la inf voi cu phap float('inf')    
    distances = {node: 0 if node == start else float('inf') for node in graph}

#tao dict voi moi node duong di
    paths = {node: [] for node in graph}
    paths[start].append(start) #luu y phai them diem start de dung logic

# ham chinh
# logic khi tim duong di dua tren nguyen tac Tham Lam, so sanh duong di moi bang tham so, neu dai hon = giu nguyen, neu ngan hon: copy + them diem cuoi
    while unvisited:
        
        #chon diem bat dau
            current = min(unvisited, key=distances.get)
            
            #so sanh de chon duong di ngan nhat
            for node, distance in graph[current]:
                
                #neu khoang cach diem moi den dich ngan hon khoang cach duong di mac dinh den dich, copy duong moi + node dich den 
                if distance + distances[current] < distances[node]:
                    distances[node] = distance + distances[current]
                    if paths[node] and paths[node][-1] == node:
                        paths[node] = paths[current][:]
                    else:
                        paths[node].extend(paths[current])
                    paths[node].append(node)
            unvisited.remove(current) # xoa lo trinh cu
            
    targets_to_print = [target] if target else graph
    for node in targets_to_print:
        if node == start:
            continue
        print(f'\n{start}-{node} distance: {distances[node]}\nPath: {" -> ".join(paths[node])}')
    
    return distances, paths
    
shortest_path(my_graph, 'A')       
```
