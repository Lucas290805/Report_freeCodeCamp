làm quen với cấu trúc dữ liệu bằng Merge_Sort
ý tưởng:
chia nhỏ các phần tử trong 1 list thành các phần tử nhỏ hơn
sắp xếp theo thứ tự khi các phần tử không thể chia nhỏ nữa

```python
def merge_sort(array):
    
    #điều kiện dừng
    if len(array) <= 1:
        return
    
    # chia nhỏ list ban đầu
    middle_point = len(array) // 2
    left_part = array[:middle_point]
    right_part = array[middle_point:]

    # Đệ quy để chia thành phần nhỏ nhất
    merge_sort(left_part)
    merge_sort(right_part)

    #Các List để lưu địa chỉ các phần tử đang tính
    left_array_index = 0
    right_array_index = 0
    sorted_index = 0

    # sort bằng cách so sánh 
    while left_array_index < len(left_part) and right_array_index < len(right_part):
        if left_part[left_array_index] < right_part[right_array_index]:
            array[sorted_index] = left_part[left_array_index]
            left_array_index += 1
        else:
            array[sorted_index] = right_part[right_array_index]
            right_array_index += 1
        sorted_index += 1

    #thêm phần tử vào sort list 
    while left_array_index < len(left_part):
        array[sorted_index] = left_part[left_array_index]
        left_array_index += 1
        sorted_index += 1
    
    while right_array_index < len(right_part):
        array[sorted_index] = right_part[right_array_index]
        right_array_index += 1
        sorted_index += 1
        
if __name__ == '__main__':
    numbers = [4, 10, 6, 14, 2, 1, 8, 5]
    print('Unsorted array: ')
    print(numbers)
    merge_sort(numbers)
    print('Sorted array: '+ str(numbers))
```
