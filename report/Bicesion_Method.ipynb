Để tìm căn bật 2 của 1 số
ta chọn 1 số nằm trong khoảng giá trị f(a) và f(b) của 1 hàm rồi dò
nếu điểm ta chọn nằm bên trái thì lấy điểm làm min hoặc ngược lại
cuối cùng thu được nghiệm gần đúng khoảng 1e-7

```python
# hàm nhận square_target = điểm cần tính. tolerance = phương sai. max_iterations = sô vòng lặp để tìm nghiệm
def square_root_bisection(square_target, tolerance=1e-7, max_iterations=100):
    
    # vì nghiệm là căn bật 2 của 1 sô nên không âm => âm là sai
    if square_target < 0:
        raise ValueError('Square root of negative number is not defined in real numbers')
    if square_target == 1:
        root = 1
        print(f'The square root of {square_target} is 1')
    elif square_target == 0:
        root = 0
        print(f'The square root of {square_target} is 0')

    else:
        low = 0
        high = max(1, square_target)# là 1 hoặc điểm chọn
        root = None
        
        for _ in range(max_iterations):
            
            # chọn 1 điểm ở giữa để bắt đầu tìm
            mid = (low + high) / 2
            square_mid = mid**2
            
            # dùng hàm abs() để lấy trị tuyệt đối, nếu bé hơn phương trai thì là nghiệm đúng
            if abs(square_mid - square_target) < tolerance:
                root = mid
                break

            elif square_mid < square_target:
                low = mid
            else:
                high = mid

        if root is None:
            print(f"Failed to converge within {max_iterations} iterations.")
    
        else:   
            print(f'The square root of {square_target} is approximately {root}')
    
    return root
```
