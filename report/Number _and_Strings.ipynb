The Luhn Algorithm

ý tưởng

- dùng để kiểm tra xem số có hợp lệ hay không bằng cách 
- cộng các số lẻ từ phải qua trái với nhân đôi các số chẳn từ phải sang trái 
- nếu số chẵn sau khi nhân >9 thì sẽ cộng 2 số lại. vd: 6x2 =12 => 1+2 = 3
- cộng tổng các số lại rồi mod 10 == 0 thì valid >< invalid

áp dụng
-các hệ thống kiểm tra tài khoản của ngân hàng

```python
def main():
    card_number = ''
    
    #loại bỏ những khoảng trắng và dấu '-'
    card_translation = str.maketrans({'-': '', ' ': ''}) 
    translated_card_number = card_number.translate(card_translation)

    if verify_card_number(translated_card_number):
        print('VALID!')
    else:
        print('INVALID!')

main()
```

```python
def verify_card_number(card_number):
    sum_of_odd_digits = 0 # tổng các số lẻ
    
    #xử lí từ phải -> trái nên phải đảo ngược trước khi xử lí
    card_number_reversed = card_number[::-1]
    odd_digits = card_number_reversed[::2] # lấy số lẻ

    for digit in odd_digits:
        sum_of_odd_digits += int(digit)

    sum_of_even_digits = 0 # tổng các số chẵn
    even_digits = card_number_reversed[1::2]# lấy số chẵn
    
    # thực hiện kiểm tra các số chẵn có >9 hay không
    for digit in even_digits:
        number = int(digit) * 2
        if number >= 10:
            number = (number // 10) + (number % 10)
        sum_of_even_digits += number
    total = sum_of_odd_digits + sum_of_even_digits
    print(total)
    return total % 10 == 0
```
